```python
import numpy as np

def reorder_phases(phases_input, ordre_actuel):
    """
    phases_input : list/np.array de longueur n (valeurs dans l'ordre "actuel")
    ordre_actuel  : liste/np.array de longueur n contenant une permutation de 1..n (ou 0..n-1)
    retourne phases réordonnées de façon naturelle (position 1 -> index 0, etc.)
    """
    phases = np.asarray(phases_input)
    ordre = np.asarray(ordre_actuel)
    n = phases.size
    if ordre.size != n:
        raise ValueError("ordre_actuel et phases doivent avoir la même longueur")
    # conversion en 0-based si nécessaire
    if ordre.min() == 1 and ordre.max() == n:
        ordre0 = ordre - 1
    elif ordre.min() == 0 and ordre.max() == n-1:
        ordre0 = ordre
    else:
        raise ValueError("ordre_actuel doit être une permutation de 1..n ou 0..n-1")
    # vérification rapide
    if set(ordre0) != set(range(n)):
        raise ValueError("ordre_actuel doit être une permutation sans répétition")
    # réordonnage (O(n))
    phases_reord = np.empty_like(phases, dtype=phases.dtype)
    phases_reord[ordre0] = phases
    return phases_reord
```
